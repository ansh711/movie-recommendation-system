'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
we have to convert it into simple list 
like --> ['action','adventure','fantasy','scifi']
now then we have to create a function using "def" by oops 

```python
import ast


def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
        
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies.head()
```

the problem is solved in genres section


```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies.head()
```

the problem is dissolved in the case of keywords

```python
def convert_cast(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!= 3 :
             L.append(i['name'])
             counter+=1
        else:
            break
              
                
            
       
    return L


#we have to have the first three main character or cast from every movie 
```

```python
movies['cast']= movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

we have to now fetch director form every movie because that is the one of the most important feature for our movie recommendation system 

```python
def fetch_dir(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director' :
             L.append(i['name'])
             break
            
       
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_dir)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x : x.split())
```

```python
movies.head()
```

if we have the spaces in every data then the duplicated data will create cluster and model will be confused ... like [Sam Worthington ] and [Sam Mendes] ....... now if we will delete thw space to convert each word the whole different tag or entity it will be very easy for the model to get the right possible answer for us and the movie recommendation system's accuracy will be enhanced 

```python
movies['genres']= movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']= movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']= movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']= movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+ movies['genres']+ movies['keywords']+ movies['cast']+ movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']= new_df['tags'].apply(lambda x: " ".join(x))
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return  " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

now we have the best possible data set or data frame for conducting machine learning algorithm on this and create or train a model for movie recommendation system 

```python
# we need to vectorize our tags using bags of words ... this is known as text vectorization 

from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
 cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
# we need to calculate the cosine distance here not the euclidean distance for the better performance 
# because the dimension is 5000 that's why in the higher dimension the euclidean distance will fail 

from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key= lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list= sorted(list(enumerate(distances)),reverse=True, key= lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
